```python
# Install xlsxwriter if not already installed
!pip install xlsxwriter

import xlsxwriter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import numpy as np # Import numpy for handling inf values

# ==========================================
# PART 1: GENERATE THE AUTO INDUSTRY EXCEL TEMPLATE
# ==========================================
filename = 'Auto_Industry_Benchmarking.xlsx'
workbook = xlsxwriter.Workbook(filename)

# --- Define Sheets ---
ws_data = workbook.add_worksheet('1. Raw_Data_Input')
ws_dash = workbook.add_worksheet('2. Analysis_Dashboard')
ws_score = workbook.add_worksheet('3. Scoring_Model')

# --- Formatting Styles ---
header_fmt = workbook.add_format({'bold': True, 'font_color': 'white', 'bg_color': '#00274C', 'border': 1, 'align': 'center', 'valign': 'vcenter'})
sub_fmt = workbook.add_format({'bold': True, 'font_color': 'black', 'bg_color': '#FFCB05', 'border': 1, 'align': 'center'})
fmt_pct = workbook.add_format({'num_format': '0.0%', 'border': 1})
fmt_usd = workbook.add_format({'num_format': '$#,##0.00', 'border': 1})
fmt_num = workbook.add_format({'num_format': '#,##0', 'border': 1})
fmt_input = workbook.add_format({'bg_color': '#EDF2F7', 'border': 1}) # Light gray for user input

# --- SHEET 1: RAW DATA INPUT (Setup for Orbis/CapitalIQ) ---
# Added 'R&D Expense' which is critical for Auto Industry analysis
headers_raw = [
    'Company Name', 'Revenue ($M)', 'COGS ($M)', 'EBITDA ($M)',
    'Net Income ($M)', 'R&D Expense ($M)', 'Total Assets ($M)', 'Total Equity ($M)', 'Employees'
]

ws_data.write('A1', 'STEP 1: PASTE ORBIS DATA BELOW', header_fmt)
ws_data.set_column('A:A', 25)
ws_data.set_column('B:I', 18)

for col, text in enumerate(headers_raw):
    ws_data.write(1, col, text, sub_fmt)

# --- INSERT DUMMY DATA (Approximate Auto Industry Figures) ---
# This ensures the template works immediately when you run it.
auto_dummy_data = [
    ['Tesla', 96000, 71000, 14000, 12000, 3000, 82000, 45000, 127000],
    ['Ford', 158000, 135000, 11000, -2000, 8000, 256000, 43000, 173000],
    ['General Motors', 156000, 130000, 14000, 9000, 9000, 264000, 68000, 167000],
    ['Toyota', 275000, 225000, 35000, 25000, 11000, 550000, 200000, 370000],
    ['Rivian', 4000, 7000, -5000, -6000, 2000, 17000, 10000, 14000],
]

for row_idx, row_data in enumerate(auto_dummy_data):
    ws_data.write(row_idx + 2, 0, row_data[0], fmt_input)
    for col_idx, val in enumerate(row_data[1:]):
        ws_data.write(row_idx + 2, col_idx + 1, val, fmt_num)

# --- SHEET 2: AUTOMATED DASHBOARD ---
dash_headers = [
    'Company Name', 'Gross Margin %', 'EBITDA Margin %',
    'Net Margin %', 'R&D Intensity (R&D/Rev)', 'ROE %', 'Rev per Employee ($)'
]

ws_dash.write('A1', 'STEP 2: AUTOMATED RATIOS', header_fmt)
ws_dash.set_column('A:A', 25)
ws_dash.set_column('B:G', 20)

for col, text in enumerate(dash_headers):
    ws_dash.write(1, col, text, sub_fmt)

# Write formulas for 20 rows
for i in range(2, 22):
    r = i + 1 # 1-based index for Excel formulas
    ws_dash.write_formula(i, 0, f"='1. Raw_Data_Input'!A{r}", fmt_input)
    # Gross Margin
    ws_dash.write_formula(i, 1, f"=IFERROR(('1. Raw_Data_Input'!B{r}-'1. Raw_Data_Input'!C{r})/'1. Raw_Data_Input'!B{r}, 0)", fmt_pct)
    # EBITDA Margin
    ws_dash.write_formula(i, 2, f"=IFERROR('1. Raw_Data_Input'!D{r}/'1. Raw_Data_Input'!B{r}, 0)", fmt_pct)
    # Net Margin
    ws_dash.write_formula(i, 3, f"=IFERROR('1. Raw_Data_Input'!E{r}/'1. Raw_Data_Input'!B{r}, 0)", fmt_pct)
    # R&D Intensity (Crucial for Auto)
    ws_dash.write_formula(i, 4, f"=IFERROR('1. Raw_Data_Input'!F{r}/'1. Raw_Data_Input'!B{r}, 0)", fmt_pct)
    # ROE
    ws_dash.write_formula(i, 5, f"=IFERROR('1. Raw_Data_Input'!E{r}/'1. Raw_Data_Input'!H{r}, 0)", fmt_pct)
    # Rev Per Employee
    ws_dash.write_formula(i, 6, f"=IFERROR('1. Raw_Data_Input'!B{r}/'1. Raw_Data_Input'!I{r} * 1000000, 0)", fmt_usd)

# --- SHEET 3: SCORING MODEL ---
ws_score.write('A1', 'STEP 3: COMPETITIVE RANKING', header_fmt)
ws_score.set_column('A:A', 25)
ws_score.write('C1', 'Weights:', sub_fmt)
ws_score.write('D1', 0.4, fmt_pct) # Weight for Margin
ws_score.write('E1', 0.4, fmt_pct) # Weight for Innovation (R&D)
ws_score.write('F1', 0.2, fmt_pct) # Weight for Efficiency

score_headers = ['Company Name', 'Margin Score (0-10)', 'Innovation Score (0-10)', 'Efficiency Score (0-10)', 'WEIGHTED SCORE']
for col, text in enumerate(score_headers):
    ws_score.write(2, col, text, sub_fmt)

# Formulas for scoring
for i in range(3, 8):
    r = i + 1
    ws_score.write_formula(i, 0, f"='1. Raw_Data_Input'!A{r}", fmt_input)
    ws_score.write(i, 1, 5, fmt_num) # Placeholder
    ws_score.write(i, 2, 5, fmt_num) # Placeholder
    ws_score.write(i, 3, 5, fmt_num) # Placeholder
    ws_score.write_formula(i, 4, f"=(B{r}*$D$1) + (C{r}*$E$1) + (D{r}*$F$1)", fmt_num)

workbook.close()
print(f"âœ… Excel File Created: {filename}")
print("   (Data populated with dummy Auto Industry figures)")


# ==========================================
# PART 2: COLAB DATA ANALYSIS & VISUALIZATION
# ==========================================
print("\nðŸ”„ Reading file back into Python for Analysis...")

# Read the Raw Data Input (Sheet 1) as the source of truth
df_raw = pd.read_excel(filename, sheet_name='1. Raw_Data_Input', header=1)

# Ensure Company Name is string
df_raw['Company Name'] = df_raw['Company Name'].astype(str)

# Calculate Dashboard metrics directly in Python
df_metrics = pd.DataFrame()
df_metrics['Company Name'] = df_raw['Company Name']

# Calculate Gross Margin %
df_metrics['Gross Margin %'] = (df_raw['Revenue ($M)'] - df_raw['COGS ($M)']) / df_raw['Revenue ($M)']
# Calculate EBITDA Margin %
df_metrics['EBITDA Margin %'] = df_raw['EBITDA ($M)'] / df_raw['Revenue ($M)']
# Calculate Net Margin %
df_metrics['Net Margin %'] = df_raw['Net Income ($M)'] / df_raw['Revenue ($M)']
# Calculate R&D Intensity (R&D/Rev)
df_metrics['R&D Intensity (R&D/Rev)'] = df_raw['R&D Expense ($M)'] / df_raw['Revenue ($M)']
# Calculate ROE %
df_metrics['ROE %'] = df_raw['Net Income ($M)'] / df_raw['Total Equity ($M)']
# Calculate Rev per Employee ($)
df_metrics['Rev per Employee ($)'] = df_raw['Revenue ($M)'] / df_raw['Employees'] * 1000000 # Convert $M to $ for rev/employee

# Clean up infinite or NaN values from division by zero if any (e.g., if Revenue or Employees is 0)
df_metrics = df_metrics.replace([np.inf, -np.inf], np.nan)
df_metrics = df_metrics.dropna(subset=df_metrics.columns.drop('Company Name'), how='all')

# Merge Revenue for bubble size into the calculated metrics dataframe
df_merged_for_plot = pd.merge(df_metrics, df_raw[['Company Name', 'Revenue ($M)']], on='Company Name')

# Ensure all necessary columns for plotting are numeric and handle potential inf/-inf/NaN from calculations
for col in ['Gross Margin %', 'R&D Intensity (R&D/Rev)', 'Revenue ($M)']:
    df_merged_for_plot[col] = pd.to_numeric(df_merged_for_plot[col], errors='coerce')
df_merged_for_plot = df_merged_for_plot.dropna(subset=['Gross Margin %', 'R&D Intensity (R&D/Rev)', 'Revenue ($M)'])

# --- VISUALIZATION: STRATEGIC GROUP MAP ---
# X-Axis: Gross Margin (Pricing Power / Brand)
# Y-Axis: R&D Intensity (Commitment to Innovation/EVs) - KEY FOR AUTO
# Size: Revenue (Market Share)

plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Create Scatter Plot
ax = sns.scatterplot(
    data=df_merged_for_plot,
    x='Gross Margin %',
    y='R&D Intensity (R&D/Rev)',
    size='Revenue ($M)',
    sizes=(300, 3000), # Range of bubble sizes
    alpha=0.7,
    hue='Company Name', # Different color for each company
    palette='deep',
    legend='full' # Ensure legend is generated by seaborn
)

# Add Labels to Bubbles
for i in range(df_merged_for_plot.shape[0]):
    row = df_merged_for_plot.iloc[i]
    plt.text(
        x=row['Gross Margin %'] + 0.002,
        y=row['R&D Intensity (R&D/Rev)'] + 0.005, # Slightly adjust y to prevent overlap with point
        s=row['Company Name'],
        fontdict={'weight': 'bold', 'size': 11} # Increased font size
    )

# Add Average Lines (Quadrants)
avg_x = df_merged_for_plot['Gross Margin %'].mean()
avg_y = df_merged_for_plot['R&D Intensity (R&D/Rev)'].mean()
plt.axvline(x=avg_x, color='gray', linestyle='--', alpha=0.5)
plt.axhline(y=avg_y, color='gray', linestyle='--', alpha=0.5)

# Annotation for Quadrants
if not df_merged_for_plot.empty:
    x_max = df_merged_for_plot['Gross Margin %'].max()
    y_max = df_merged_for_plot['R&D Intensity (R&D/Rev)'].max()
    x_min = df_merged_for_plot['Gross Margin %'].min()
    y_min = df_merged_for_plot['R&D Intensity (R&D/Rev)'].min()

    plt.text(x_max * 0.95, y_max * 0.95, # Adjust position slightly
             'High Innovation\nHigh Margin', color='green', ha='right', va='top', fontsize=12)
    plt.text(x_min * 1.05, y_min * 1.05, # Adjust position slightly
             'Commoditized\nLegacy', color='red', ha='left', va='bottom', fontsize=12)

# Formatting
plt.title('Auto Industry Strategic Group Map: Innovation vs. Pricing Power', fontsize=18, weight='bold') # Increased title font size
plt.xlabel('Gross Margin % (Proxy for Premium Brand/Pricing)', fontsize=14) # Increased x-label font size
plt.ylabel('R&D Intensity (R&D % of Revenue)', fontsize=14) # Increased y-label font size

# Format axes as Percentages
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:.0%}'.format(x)))
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:.0%}'.format(y)))

# Customizing legend (Seaborn creates it, we just position it)
if ax.get_legend(): # Check if legend was actually created
    plt.setp(ax.get_legend().get_texts(), fontsize='10') # Adjust legend item text size
    plt.setp(ax.get_legend().get_title(), fontsize='12') # Adjust legend title size
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.) # Move legend outside to prevent overlap

plt.tight_layout()
plt.show()

print("\nâœ… Visualization Complete.")
```

```python
# Helper function to normalize scores to a 0-10 scale, handling cases with constant values
def normalize_to_0_10(series):
    min_val = series.min()
    max_val = series.max()
    if max_val == min_val:
        return pd.Series([5.0] * len(series), index=series.index) # Assign 5 if all values are the same
    return 10 * (series - min_val) / (max_val - min_val)

# Create a new DataFrame for scoring, similar to the Excel '3. Scoring_Model' structure
df_scoring = pd.DataFrame()
df_scoring['Company Name'] = df_metrics['Company Name']

# Calculate Margin Score (based on Gross Margin %)
df_scoring['Margin Score (0-10)'] = normalize_to_0_10(df_metrics['Gross Margin %'])

# Calculate Innovation Score (based on R&D Intensity (R&D/Rev))
df_scoring['Innovation Score (0-10)'] = normalize_to_0_10(df_metrics['R&D Intensity (R&D/Rev)'])

# Calculate Efficiency Score (based on Rev per Employee ($))
df_scoring['Efficiency Score (0-10)'] = normalize_to_0_10(df_metrics['Rev per Employee ($)'])

# Calculate WEIGHTED SCORE
df_scoring['WEIGHTED SCORE'] = (
    df_scoring['Margin Score (0-10)'] * margin_weight +
    df_scoring['Innovation Score (0-10)'] * innovation_weight +
    df_scoring['Efficiency Score (0-10)'] * efficiency_weight
)

# Display the results
print("\nCalculated Weighted Scores:")
display(df_scoring.round(2))
```
